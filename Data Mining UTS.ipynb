```python
  #import libraly pandas, seaborn, matplotlib
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

path ="https://raw.githubusercontent.com/Moonrack/uts/main/loan_data.csv"
```

```python
pd.set_option( 'display.max_columns' , None )
```

```python
#review data dari link path
df=pd.read_csv(path)
df.head(100)
```

```python
#cek duplikasi data
df.duplicated().sum()
```

```python
#Mendroop / menghapus Columns Loan_ID
df.drop(columns=['Loan_ID'],inplace=True)
df.head()
```

```python
#mengechek data kosong
df.isna()
```

```python
#cek berapa data kosong yang ada di columns
df.isna().sum()
```

```python
baris,kolom= df.shape
print (f'jumlah data ada {baris} record, jumlah fitur ada: {kolom} kolom')
```

```python
#menghapus kolom data yang paling bangak kosongnya
df.drop(columns=['Credit_History'], inplace=True)
df.head()
```

```python
#menampilkan sample data berjumlah 10 dengan no acak secara random
df.sample(10)
```
