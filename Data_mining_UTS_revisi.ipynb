```python
  #import libraly pandas, seaborn, matplotlib
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

path ="https://raw.githubusercontent.com/Moonrack/uts/main/loan_data.csv"
```

```python
pd.set_option( 'display.max_columns' , None )
```

```python
#review data dari link path
df=pd.read_csv(path)
df.head(100)
```

```python
#cek duplikasi data
df.duplicated().sum()
```

```python
#Mendroop / menghapus Columns Loan_ID
df.drop(columns=['Loan_ID'],inplace=True)
df.head()
```

```python
#mengechek data kosong
df.isna()
```

```python
#cek berapa data kosong yang ada di columns
df.isna().sum()
```

```python
baris,kolom= df.shape
print (f'jumlah data ada {baris} record, jumlah fitur ada: {kolom} kolom')
```

```python
#menghapus kolom data yang paling bangak kosongnya
df.drop(columns=['Credit_History'], inplace=True)
df.head()
```

```python
#cek berapa data kosong yang ada di columns
df.isna().sum()
```

```python
#Mengisi data kategori dengan data terbanyak
df.Self_Employed.fillna('S', inplace=True)
df.isna().sum()
```

```python
df.Loan_Amount_Term.fillna('S', inplace=True)
df.isna().sum()
```

```python
#Mengisi data kategori dengan data terbanyak
df.Gender.fillna('Female', inplace=True)
df.isna().sum()
```

```python
#Menampilkan data kolom "Gender" dan ternyata dominan male = Laki-laki
sns.displot(data= df, x= 'Gender',bins=30, kde=True)
```

```python
# Menghitung jumlah peminjam berdasarkan gender
gender_count = df['Gender'].value_counts()
```

```python
# Menampilkan hasil
print(gender_count)
```

```python
# Menampilkan dalam bentuk grafik
gender_count.plot(kind='bar', color=['blue', 'orange'])
plt.title('Jumlah Peminjam Berdasarkan Gender')
plt.xlabel('Gender')
plt.ylabel('Jumlah Peminjam')
plt.xticks(rotation=0)
plt.show()
```

```python
# Filter data hanya untuk laki-laki
male_df = df[df['Gender'] == 'Male']
```

```python
male_count_by_area = male_df['Property_Area'].value_counts()
```

```python
# Menampilkan hasil
print(male_count_by_area)
```

```python
male_count_by_area.plot(kind='bar', color=['blue', 'orange', 'green'])
plt.title('Jumlah Laki-laki Berdasarkan Property Area')
plt.xlabel('Property Area')
plt.ylabel('Jumlah Laki-laki')
plt.xticks(rotation=0)
plt.show()
```

```python
# Filter data hanya untuk laki-laki yang tinggal di kotaFilter data hanya untuk laki-laki yang tinggal di kota
male_city_df = df[(df['Gender'] == 'Male') & (df['Property_Area'] == 'Urban')]
```

```python
# Menghitung jumlah laki-laki di kota berdasarkan status pekerjaan
employment_status_count = male_city_df['Self_Employed'].value_counts()
```

```python
# Menampilkan hasil
print(employment_status_count)
```

```python
#menampilkan grafik Status Pekerjaan Laki-laki di Kota
employment_status_count.plot(kind='bar', color=['blue', 'orange'])
plt.title('Status Pekerjaan Laki-laki di Kota')
plt.xlabel('Self_Employed')
plt.ylabel('Jumlah')
plt.xticks(rotation=0)
plt.show()
```

```python
# Filter data hanya untuk laki-laki yang bekerja sendiri dan tinggal di kota
male_city_self_employed_df = df[(df['Gender'] == 'Male') &
                                (df['Property_Area'] == 'Urban') &
                                (df['Self_Employed'] == 'Yes')]
```

```python
# Menghitung jumlah laki-laki di kota berdasarkan status pendidikan
education_status_count = male_city_self_employed_df['Education'].value_counts()
```

```python
# Menampilkan hasil
print(education_status_count)
```

```python
#Menampilkan Grafik Status Pendidikan Laki-laki Pekerja di Kota
education_status_count.plot(kind='bar', color=['blue', 'orange'])
plt.title('Status Pendidikan Laki-laki Pekerja di Kota')
plt.xlabel('Education')
plt.ylabel('Jumlah')
plt.xticks(rotation=0)
plt.show()
```

```python


# Menampilkan pinjaman terendah dan tertinggi
loan_amount_min = male_city_self_employed_df['LoanAmount'].min()
loan_amount_max = male_city_self_employed_df['LoanAmount'].max()

print(f"Pinjaman terendah: {loan_amount_min}")
print(f"Pinjaman tertinggi: {loan_amount_max}")
```

```python
# Filter data hanya untuk laki-laki yang tinggal di kota dan berstatus Not Graduate
male_city_not_graduate_df = df[(df['Gender'] == 'Male') &
                               (df['Property_Area'] == 'Urban') &
                               (df['Education'] == 'Not Graduate')]
```

```python
# Menghitung jumlah laki-laki di kota berdasarkan jumlah tanggungan
dependents_status_count = male_city_not_graduate_df['Dependents'].value_counts()
```

```python
# Menampilkan hasil
print(dependents_status_count)
```

```python

#Menampilkan Grafik Status Tanggungan Laki-laki di Kota yang Not Graduate
dependents_status_count.plot(kind='bar', color=['blue', 'orange', 'green', 'red'])
plt.title('Status Tanggungan Laki-laki di Kota yang Not Graduate')
plt.xlabel('Dependents')
plt.ylabel('Jumlah')
plt.xticks(rotation=0)
plt.show()
```
